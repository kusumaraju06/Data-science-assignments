```python
# Calculate and print missing values for each column
missing_values = data.isnull().sum()
print(missing_values)

# Define a function to get fill values based on data type
def get_fill_value(series):
    if series.dtype == 'object':
        return series.mode()[0]
    else:
        return series.mean()

# Fill missing values in each column based on its type
for column in data.columns:
    if missing_values[column] > 0:  # Check if there are missing values
        fill_value = get_fill_value(data[column])
        data[column].fillna(fill_value, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler_standard = StandardScaler()
data_standard_scaled = data.copy()
data_standard_scaled[data_standard_scaled.select_dtypes(include=['int64', 'float64']).columns] = scaler_standard.fit_transform(data_standard_scaled.select_dtypes(include=['int64', 'float64']))
data_standard_scaled
```

```python
scaler_minmax = MinMaxScaler()
data_minmax_scaled = data.copy()
data_minmax_scaled[data_minmax_scaled.select_dtypes(include=['int64', 'float64']).columns] = scaler_minmax.fit_transform(data_minmax_scaled.select_dtypes(include=['int64', 'float64']))
data_minmax_scaled
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Identify categorical columns with fewer than 5 unique values
categorical_columns = data.select_dtypes(include=['object']).nunique()[lambda x: x < 5].index.tolist()

# Initialize OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse_output=False, drop='first')

# Apply OneHotEncoder to the selected categorical columns
encoded_data = one_hot_encoder.fit_transform(data[categorical_columns])

# Create a DataFrame with the encoded features
encoded_df = pd.DataFrame(encoded_data, columns=one_hot_encoder.get_feature_names_out(categorical_columns))

# Concatenate the remaining data with the one-hot encoded features
data_one_hot_encoded = pd.concat([data.drop(columns=categorical_columns), encoded_df], axis=1)

# Display the final DataFrame with one-hot encoded features
data_one_hot_encoded

```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to object-type columns that are not in categorical_columns
for col in data.select_dtypes(include=['object']).columns:
    if col not in categorical_columns:
        data[col] = label_encoder.fit_transform(data[col])

# Display the label encoder object
label_encoder
```
